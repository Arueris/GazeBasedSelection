```python
import pandas as pd
import os

path = "HyperparameterResults_LSTM_fps90"

results = pd.read_csv(os.path.join(path, "results_log.csv"))
results
```

```python
cond = "gaze"
for cond in results.cond.unique():
    print("="*80)
    print(" "*30, cond)
    best_correct = results.loc[results.cond==cond, "correct_acc"].max()
    best_incorrect = results.loc[results.cond==cond, "incorrect_acc"].max()

    best_correct_models = results.loc[(results.correct_acc==best_correct) & (results.cond==cond)]
    best_incorret_models = results.loc[(results.incorrect_acc==best_incorrect) & (results.cond==cond)]

    print("="*80)
    print("Best Correct")
    display(best_correct_models)
    print()
    print()
    print("-"*80)
    print("Best Incorrect")
    display(best_incorret_models)
    print()
    print()
```
